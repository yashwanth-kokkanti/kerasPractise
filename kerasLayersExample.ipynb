```
## This is a simple note book to practise Keras Layers 

## Imports 

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import plot_model 

```

```
## define Sequential Layer 
model = Sequential()

model.add(Dense(16, input_shape=(8, )))
model.summary()

plot_model(model, show_shapes=True)
```

```
model.get_config()
```

```
model.get_weights()
```

```
### Example of Dropout Layers

from tensorflow.keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(16, input_shape=(8, )))
model.add(Dropout(0.2))
model.add(Dense(10))
model.summary()

plot_model(model, show_shapes=True)
```

```
## Here Layer_1 params explained are input = 8 and output = 16 total params = 8 * 16  + 16 (Bias terms) = 128 + 16 = 144
## Layer_2 = Params input = 16 and output = 10 and bias terms = 10 : 16 * 10 + 10 = 170
```

```
## Flatten Layer => Flatten is used to flatten the output data to single dimension. 

from tensorflow.keras.layers import Dense, Flatten

model = Sequential()
model.add(Dense(16, input_shape=(8, 2)))
model.add(Flatten())
model.add(Dense(10))
model.summary()
plot_model(model, show_shapes=True, show_layer_names=True)
```

```
## Permute layer used to change the shape of the input using a specific pattern

from tensorflow.keras.layers import Dense, Permute

model  = Sequential()
model.add(Dense(16, input_shape=(8, 2)))
model.add(Permute((2,1 )) )

model.summary()
```

```
##  Conv2D Layer Example 

from tensorflow.keras.layers import Dense, Activation, Conv2D, MaxPooling2D, Flatten

model = Sequential()

model.add(Conv2D(32, kernel_size=(5, 5), strides=(1,1), input_shape=(28, 28, 1)))

model.add(MaxPooling2D(pool_size=(2,2), strides=(2, 2)))

model.add(Flatten())
model.add(Dense(1000, activation='relu'))

plot_model(model, show_shapes=True, show_layer_names=True)
```
